```python
#Import QTLseqr package into python environment
QTLseqr = importr("QTLseqr")
```

```python
# VCF file must only contain biallelic variants. (Filter upstream, e.g., with bcftools view -m2 -M2 )
vcf =  vcfR.read_vcfR(file = "wGQ-Filt-freebayes~bwa~IRGSP-1.0~both-segregant_bulks~filtered-default.vcf.gz")
```

```python
#Convert to Tidy Data Frame
VCF_TIDY = vcfR.vcfR2tidy(vcf)
```

```python
# Call the Parser
QTLseqr.QTLParser_1_MH(vcf = VCF_TIDY, HighBulk = "ET-pool-385", LowBulk = "ES-pool-430")
```

```python
#Import pandas and rpy2 package
import pandas as pd 
import rpy2.robjects as ro
from rpy2.robjects import pandas2ri
from rpy2.robjects.conversion import localconverter

#Set High bulk and Low bulk sample names and parser generated file name
HighBulk = "ET-pool-385"
LowBulk = "ES-pool-430"
file = "Hall.csv"

#Make a list of Chromosome Values
Chroms = list(range(1,12+1))
print(Chroms)

#Import the CSV file with importFromTable function
df = QTLseqr.importFromTable(file = file,highBulk = HighBulk,lowBulk = LowBulk,chromList = Chroms)

#Covert R Dataframe to Pandas Dataframe
with localconverter(ro.default_converter + pandas2ri.converter):
    df2 = ro.conversion.rpy2py(df)

```

```python
#Inspect the contents of dataframe
df2.info()
```

```python
# Import pandas numpy matplotlib and plotnine packages 
import pandas as pd
import numpy as np

import plotnine
from plotnine import *
from plotnine.data import *

%matplotlib inline

#Define a variable x as the sum of Alternate Alleles called per segregated bulk
x = df2['AD_ALT.LOW'] + df2['AD_ALT.HIGH']

#Plot a histogram of x
ggplot(df2, aes(x=x)) + geom_histogram(bins = 150) + coord_cartesian(xlim=(0, 400), ylim=(0, 200000)) + plotnine.labels.ggtitle('Total Alternate Allelic Depth (AD_ALT.LOW + AD_ALT.HIGH)')

```

```python
#Define a variable x as the sum of Reference Alleles called per segregated bulk
x = df2['AD_REF.LOW'] + df2['AD_REF.HIGH']
#Plot it
ggplot(df2, aes(x=x)) + geom_histogram(bins = 500) + coord_cartesian(xlim=(0, 400)) + plotnine.labels.ggtitle('Total Reference Allelic Depth (AD_REF.LOW + AD_REF.High)')


```

```python
#Define a variable x as the total sum of Alleles called
x = df2['DP.LOW'] + df2['DP.HIGH']
#Plot it is a histogram
ggplot(df2, aes(x=x))+ geom_histogram(bins = 500) + coord_cartesian(xlim=(0, 400)) + plotnine.labels.ggtitle('Total Depth')

```

```python

#Plot Reference Frequency 
ggplot(df2, aes(x='REF_FRQ')) + geom_histogram(bins = 50) + plotnine.labels.ggtitle('Reference Allelic Frequency')
```

```python
# Use Histograms to justify filtering cutoff values
# Call filterSNPs functions
df_filt = QTLseqr.filterSNPs(SNPset = df,refAlleleFreq = 0.20,minTotalDepth = 100,maxTotalDepth = 400,minSampleDepth = 40)
```

```python
#Run G' analysis
df_filt = QTLseqr.runGprimeAnalysis_MH(SNPset = df_filt,windowSize = 5000000,outlierFilter = "deltaSNP",filterThreshold = 0.1)
```

```python
#Run QTL Sequence Analysis Function
vec1 = 430,385
R_float_vec1 = ro.vectors.FloatVector(vec1)
vec2 = 95,99
R_float_vec2 = ro.vectors.FloatVector(vec2)

df_filt2 = QTLseqr.runQTLseqAnalysis_MH(SNPset = df_filt,windowSize = 5000000,popStruc = "F2",bulkSize = R_float_vec1,replications = 10000,intervals = R_float_vec2)
```

```python
#Plot GPrime Distribution and Estimated Null Distribution Parameters
print(QTLseqr.plotGprimeDist_MH(SNPset = df_filt2, outlierFilter = "Hampel"))
```

```python
#Plot G Prime Distribution with deltaSNP outlierfilter
print(QTLseqr.plotGprimeDist_MH(SNPset = df_filt2, outlierFilter = "deltaSNP",filterThreshold = 0.1))
```

```python
#Plot the number of SNPs
print(QTLseqr.plotQTLStats(SNPset = df_filt2, var = "nSNPs"))
```

```python
#Plot Gprime statistic per chromosome
print(QTLseqr.plotQTLStats(SNPset = df_filt, var = "Gprime", plotThreshold = True, q = 0.01))
```

```python
#Plot deltaSNP Statistic
print(QTLseqr.plotQTLStats(SNPset = df_filt2, var = "deltaSNP", plotIntervals = True))
```

```python
#Plot negLog10Pval
print(QTLseqr.plotQTLStats(SNPset = df_filt2, var = "negLog10Pval",plotThreshold = True,q=0.01))
```
