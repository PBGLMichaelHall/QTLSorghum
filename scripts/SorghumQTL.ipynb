```python
# VCF file must only contain biallelic variants. (Filter upstream, e.g., with bcftools view -m2 -M2 )
vcf =  vcfR.read_vcfR(file = "freebayes_D2.filtered.vcf")
```

```python
#Convert to Tidy Data Frame
VCF_TIDY = vcfR.vcfR2tidy(vcf)
```

```python
# Call the Parser
QTLseqr.QTLParser_1_MH(vcf = VCF_TIDY, HighBulk = "D2_F2_tt", LowBulk = "D2_F2_TT")
```

```python
import pandas as pd 
import rpy2.robjects as ro
from rpy2.robjects import pandas2ri
from rpy2.robjects.conversion import localconverter

#Set High bulk and Low bulk sample names and parser generated file name
HighBulk = "D2_F2_tt"
LowBulk = "D2_F2_TT"
file = "Hall.csv"


Chroms = list(range(1,10+1))
df = QTLseqr.importFromTable(file = file,highBulk = HighBulk,lowBulk = LowBulk,chromList = Chroms)
#plot histograms associated with filtering arguments to determine if cut off values are appropriate
with localconverter(ro.default_converter + pandas2ri.converter):
    df2 = ro.conversion.rpy2py(df)


```

```python
# Use Histograms to justify filtering cutoff values
df_filt = QTLseqr.filterSNPs(SNPset = df,refAlleleFreq = 0.20,minTotalDepth = 100,maxTotalDepth = 400,minSampleDepth = 40)
```

```python
#Run G' analysis
df_filt = QTLseqr.runGprimeAnalysis_MH(SNPset = df_filt,windowSize = 5000000,outlierFilter = "deltaSNP",filterThreshold = 0.1)
```

```python
vec1 = 45,38
R_float_vec1 = ro.vectors.FloatVector(vec1)
vec2 = 95,99
R_float_vec2 = ro.vectors.FloatVector(vec2)

df_filt2 = QTLseqr.runQTLseqAnalysis_MH(SNPset = df_filt,windowSize = 5000000,popStruc = "F2",bulkSize = R_float_vec1,replications = 10000,intervals = R_float_vec2)
```
